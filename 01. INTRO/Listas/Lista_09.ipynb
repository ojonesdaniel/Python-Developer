1. Segundo Maior Elemento
Enunciado:
Faça um programa que receba um vetor de números inteiros e encontre o segundo maior
elemento.


2. Rotação de Vetor à Direita
Enunciado:
Rotacione um vetor uma posição à direita. O último elemento deve ir para a primeira
posição.


3. Contar Frequência de um Número
Enunciado:
Leia um vetor e um número X. Conte quantas vezes X aparece no vetor.


4. União de Vetores (sem repetição)
Enunciado:
Leia dois vetores e crie um terceiro contendo a união dos elementos, sem valores
duplicados.

5. Separar Pares e Ímpares
Enunciado:
Leia um vetor de inteiros e separe os números pares e ímpares em dois novos vetores.
