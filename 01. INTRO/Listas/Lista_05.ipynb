3. Calcular Média de Notas: Peça ao usuário para inserir as notas de um
conjunto de alunos. Utilize um loop para permitir a inserção de notas até que
o usuário deseje parar. Em seguida, calcule e exiba a média das notas.

```python
lista_notas = list()
nota = int(input("Digite a nota:"))

lista_notas.append(nota)


while nota != -1:
    nota = int(input("Digite uma nova nota:"))
    print("Para encerrar a incerção de notas digite -1")
    lista_notas.append(nota)
else:
    print("Inserção encerrada.")

if sum(lista_notas) > 0:
    media = sum(lista_notas) / len(lista_notas)
    print(f"A média das notas digitadas é igual a {media}")
else:
    print("Não foram digitados valores válidos.")
```
