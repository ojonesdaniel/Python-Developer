1. Maior de 3 Números:

● Leia 3 números (a, b, c)

● Exiba o maior número entre eles


```python
num_01 = int(input("Digite o primeiro número:"))
num_02 = int(input("Digite o segundo número:"))
num_03 = int(input("Digite o terceiro número:"))

numeros = [num_01, num_02, num_03]

print(f"O maior número dentro os valores digitados é: {max(numeros)}")
```

2. Positivo ou Negativo:

● Leia um número (n)

● Exiba se o número é positivo ou negativo


```python
num = int(input(f"Digite um número"))


if num > 0:
  print(f"Valor é um número positivo")
else:
  print(f"Valor é um número negativo")
```

3. Par ou Ímpar:

● Leia um número (n)

● Exiba se o número é par ou ímpar


```python
num = int(input(f"Digite um número"))


if num % 2 <  1:
  print(f"Valor é um número par")
else:
  print(f"Valor é um número ímpar")
```

4. Cálculo de IMC:

● Leia o peso (kg) e a altura (m) de uma pessoa

● Calcule o IMC (IMC = peso / altura²)

● Exiba a classificação do IMC:

○ Abaixo de 18.5: Abaixo do peso

○ Entre 18.5 e 24.9: Peso normal

○ Entre 25.0 e 29.9: Excesso de peso

○ Acima de 30.0: Obesidade

```python
nome = input("Qual é o seu nome?")
altura = float(input("Qual é a sua altura?"))
peso = float(input( "Qual é o seu peso?"))

imc = peso/(altura**2)

print("Olá {}! com a altura {}m e o peso {}kg, seu IMC é {}." .format(nome, altura, peso, imc)) #com format
#print(f"Olá {nome}! com a altura {altura}m e o peso {peso}kg, seu IMC é {imc}.") #com f-string

if 25 < imc < 29.9:
    print("Sobrepeso")
elif 18.5 < imc < 24.9:
    print("Peso normal")
elif imc < 18.5:
    print("Abaixo do peso normal")
else:
    print("Obesidade")
```

5. Calculadora de Desconto:

● Leia o valor original de um produto

● Leia o valor do desconto (em porcentagem)

● Calcule o valor final com o desconto aplicado

● Exiba o valor original, o valor do desconto e o valor final


```python
valor_produto = float(input(f"Digite o valor do produto:"))
perc_desconto = int(input(f"Digite o percentual de desconto:"))

valor_desconto = (valor_produto / 100) * perc_desconto
valor_a_pagar = (valor_produto / 100) * (100 - perc_desconto)

print(f"O valor original do produto é: {valor_produto}")
print(f"O desconto do produto é de: {perc_desconto}% que seria R${valor_desconto}")
print(f"O valor do produto a ser pago: R${valor_a_pagar}")
```
