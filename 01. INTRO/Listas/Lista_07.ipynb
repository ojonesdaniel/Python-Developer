```python
valores = list()

while True:
    num = int(input("Digite um número para tirar a média e o valor -5 para encerrar a lista"))
    if num != -5:
        valores.append(num)
    else:
        break

soma = sum(valores)
quantidade = len(valores)

if len(valores) > 0:
    media = soma / quantidade
    print(f"A média dos valores digitados é igual a {media}.")
    print(f"O valores digitados anteriomente são: {valores}")
else:
    print("Nenhum valor válido foi informado")
```

```python
valores = []

valor = int(input("Digite um número (-5 para sair): "))


while valor != -5:
    valor = int(input("Digite um número (-5 para sair): "))
    valores.append(valor)
else:
    print("Sem valores válidos.")


if len(valores) > 0:
    print(f"Esse é o maior dos números: {max(valores)}")
    print(f"O menor número é {min(valores)}")
    
    valores_maiores_10 = [v for v in valores if v > 10]
    print(f"Esses são os números maiores que 10: {valores_maiores_10}")
else:
    print("Nenhum valor válido foi digitado.")
```
