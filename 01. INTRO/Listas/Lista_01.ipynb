11. Soma dos pares

Peça ao usuário 10 números e mostre a soma apenas dos números pares.


```python
num_1 = int(input("Digite o primeiro número:"))
num_2 = int(input("Digite o  segundo número:"))
num_3 = int(input("Digite o  terceiro número:"))
num_4 = int(input("Digite o  quarto número:"))
num_5 = int(input("Digite o quinto número:"))
num_6 = int(input("Digite o sexto número:"))
num_7 = int(input("Digite o setimo número:"))
num_8 = int(input("Digite o oitavo número:"))
num_9 = int(input("Digite o nono número:"))
num_10 = int(input("Digite o decimo número:"))

lista_num = [num_1, num_2, num_3, num_4, num_5, num_6, num_7, num_8, num_9, num_10]

print(f"Seus números são: {lista_num}")

lista_pares = list()


print("Validando números pares...")
for numero in lista_num:
  if numero % 2 == 0:
    lista_pares.append(numero)

print(f"Números Pares: {lista_pares}")

soma_pares = sum(lista_pares)

print(f"A soma dos valores pares é igual a {soma_pares}")
```

12. Fatorial

Leia um número inteiro positivo e calcule seu fatorial.


```python
import math

num_1 = int(input("Digite um número: "))

lista_fatorial = []

#Use for: quando você já sabe quantas vezes vai repetir.
#Use while: quando você não sabe exatamente quantas vezes, e quer repetir enquanto uma.

while num_1 >= 1:
    lista_fatorial.append(num_1)
    num_1 -= 1

print(f"Sua lista de números para encontrar o Fatorial é:{lista_fatorial}")

resultado = math.prod(lista_fatorial)

print(f"Seu resultado é: {resultado}")
```

13. Sequência de Fibonacci

Mostre os 15 primeiros termos da sequência de Fibonacci.


```python
# N.T
n = int(input("Quantos termos da sequência de Fibonacci você quer? "))

fibonacci = []

a, b = 0, 1

for _ in range(n):
    fibonacci.append(a)
    a, b = b, a + b

print(fibonacci)
```

15. Vetor e média

Leia 5 números e armazene em um vetor.
 Depois:

*	Mostre todos os números
* Calcule a média
*	Mostre o maior valor


```python
nums_vetor = list(map(int, input("Digite 5 valores separados por espaço: ").split()))

print(f"Esses são seus valores: {nums_vetor}")
print(f"A média desses valores é: {sum(nums_vetor) / len(nums_vetor):.2f}")
print(f"O maior desses valores é: {max(nums_vetor)}")
```

16. Jogo da adivinhação
O programa deve gerar um número aleatório entre 1 e 100 e o usuário deve tentar adivinhar.


```python
import random

num_aleatorio = random.randint(1, 10)
num_escolhido = int(input("Digite um número:"))
print(num_aleatorio)

while num_escolhido != num_aleatorio:
    print("Você errou!!")
    num_escolhido = int(input("Digite outro número:"))
else:
    print(f"Você acertou!!!! o número era {num_aleatorio}")
```

17. Sistema de notas
Leia o nome de 5 alunos e suas respectivas notas.
 Mostre:

* Média da turma
* Nome do aluno com maior nota


```python
alunos = list(map(str, input("Digite o nome dos alunos (separados por espaço): ").split()))
notas = list(map(int, input("Digite as 5 notas (separados por espaço): ").split()))

media = sum(notas) / len(notas)

maior_nota = max(notas)
indice_maior_nota = notas.index(maior_nota)
maior_nota_aluno = alunos[indice_maior_nota]

print(f"A média da turma foi {media}")
print(f"A maior nota da sala foi {maior_nota} do aluno {maior_nota_aluno}")
```
