1.Leia três números inteiros e imprima a média aritmética entre esses números.

```python
num_1 = int(input("Digite o primeiro número:"))
num_2 = int(input("Digite o segundo número:"))
num_3 = int(input("Digite o terceiro número:"))

lista_nums = [num_1, num_2, num_3]

num_media = sum(lista_nums)
den_media = len(lista_nums)

media = num_media / den_media

print(f"O Valor da média é: {media}")
```

2.Faça um programa que receba o ano de nascimento de uma pessoa, o ano atual e imprima:
A idade da pessoa no ano atual
A idade que a pessoa terá em 2050

```python
from datetime import date

ano_nascimento = int(input("Digite o ano do seu nascimento:"))
ano_escolhido = int(input("Digite o ano qualquer do futuro:"))
ano_atual = date.today().year

idade_atual = ano_atual - ano_nascimento
idade_escolhida = ano_escolhido - ano_nascimento


print(f"Sua idade atual é {idade_atual} anos.")
print(f"Sua idade no ano de {ano_escolhido} é {idade_escolhida} anos.")

```

3.Faça um programa que receba a cotação do dólar em reais, e um valor que o usuário possui em dólares. Imprima este valor em reais.

5.Faça um programa que receba uma temperatura em Celsius, calcule e mostre essa temperatura em Fahrenheit (F = (C*1,8) + 32)

```python
celsius = float(input("Digite a temperatura em graus Celsius:"))

fahrenheit = (celsius * 1.8) + 32
fahrenheit = round(fahrenheit, 2)

print(f"A temperatura em Fahrenheit é de {round(fahrenheit, 2)}ºF.")
```

6.Faça um programa que solicite ao usuário que informe os coeficientes a, b e c de uma equação de segundo grau, e que imprima as raízes desta equação (considere que os valores informados sempre retornarão raízes reais para a equação).

```python
a = float(input("Digite o coeficiente A:"))
b = float(input("Digite o coeficiente B:"))
c = float(input("Digite o coeficiente C:"))


delta = b**2 - (4 * a * c)

x = (-b - (delta ** 0.5) )/(2 * a)

print(f"O valor de x é igual a: {x}")
```

9.Receba um número positivo, calcule e mostre:

- O número digitado ao quadrado

- O número digitado ao cubo

- A raiz quadrada do número digitado

- A raiz cúbica do número digitado.

```python

num = int(input("Digite um número para fazermos as operações:"))

ao_quadrado = num ** 2
ao_cubo = num ** 3

raiz_quadrada = num ** (1/2)
raiz_cubica = num ** (1/3)

print(f"O número {num} ao quadrado é {ao_quadrado}.")
print(f"O número {num} ao cubo é {ao_cubo}.")
print(f"O número {num} tem raiz quadrada no valor de {raiz_quadrada}.")
print(f"O número {num} tem raiz cubica no valor de {raiz_cubica}.")
```

14.Faça um programa que leia duas variáveis e troque o conteúdo dessas duas variáveis. Em seguida, imprima o valor dessas variáveis invertido. Exemplo: A = 7, B = 9. Saída: A = 9, B = 7.

```python
a = int(input("Digite um número:"))
b = int(input("Digite um segundo número:"))

print(f"o valor de A é {a} e o valor de B é {b}")

print("Invertendo os valores...")
a = a + b # a passa a ser a soma dos dois (10 + 5 = 15)
b = a - b # b recebe a soma - b original (15 - 5 = 10) -> b agora é 10
a = a - b # a recebe a soma - novo b (15 - 10 = 5) -> a agora é 5

print(f"o valor de A é {a} e o valor de B é {b}")
```

15.Considerando uma eleição de apenas dois candidatos, faça um programa que leia o número total de eleitores, o número de votos do primeiro candidato e o número de votos do segundo candidato.

Em seguida, o programa deverá apresentar o percentual de votos de cada um dos candidatos e o percentual de votos nulos.

```python
votos_lula = int(input("Digite o número de votos que o Lula recebeu:"))
votos_bozo = int(input("Digite o número de votos que o Bolsonaro recebeu:"))
votos_nulos = int(input("Digite o número de votos nulos:"))


total_votos = votos_lula + votos_bozo + votos_nulos

perc_lula = round((votos_lula / total_votos) * 100, 2)
perc_bozo = round((votos_bozo / total_votos) * 100, 2)
perc_nulo = round((votos_nulos / total_votos)* 100, 2)

print(f"O percentuais ficaram:")
print(f"Lula: {perc_lula}%")
print(f"Bolsonaro: {perc_bozo}%")
print(f"Nulo: {perc_nulo}%")
```
