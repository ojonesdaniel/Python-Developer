5. Conversor de temperatura
Leia uma temperatura em Celsius e converta para Fahrenheit.
 Fórmula: F = (C * 9/5) + 32


```python
celsius = input("Digite sua temperatura:")

celsius = float(celsius)

fahrenheit = (celsius * 9/5) + 32

fahrenheit = float(fahrenheit)

print(f"Sua temperatura em Fahrenheit é: {fahrenheit}ºF")
```

6. Par ou Ímpar
Leia um número inteiro e informe se ele é par ou ímpar.


```python
num_1 = input("Digite um número: ")

if num_1.endswith(("1", "3", "5", "7", "9")):
    print(f"O {num_1} é número ímpar.")
else:
    print(f"O {num_1} é número par.")
```

7. Maior número
Leia três números e informe qual é o maior.


```python
num_1 = int(input("Digite um número: "))
num_2 = int(input("Digite um número: "))
num_3 = int(input("Digite um número: "))

lista = [num_1, num_2, num_3]
maior_num = max(lista)

print(f"O maior número é {maior_num}")
```
