1. A padaria Hotpão vende uma certa quantidade de pães franceses e uma quantidade de broas a cada dia. Cada pãozinho custa R$ 0,12 e a broa custa R$ 1,50. Ao final do dia, o dono quer saber quanto arrecadou com a venda dos pães e broas (juntos), e quanto deve guardar numa conta de poupança (10% do total arrecadado). Você foi contratado para fazer os cálculos para o dono. Com base nestes fatos, faça um algoritmo para ler as quantidades de pães e de broas, e depois calcular os dados solicitados.


```python
valor_pao = 0.12
valor_broa = 1.5

qtde_pao = int(input("Quantidade de pães vendidos no dia de hoje:"))
qtde_broa = int(input("Quantidade de broas vendidos no dia de hoje:"))


arrecadacao = (valor_pao * qtde_pao) + (valor_broa * qtde_broa)
poupanca = (0.1 * arrecadacao)

print(f"O valor arrecado no dia é de: R${arrecadacao}")
print(f"Valor a ser adicionado na poupança é de? R${poupanca}")
```


2. Escreva um algoritmo para ler o nome e a idade de uma pessoa, e exibir quantos dias de vida ela possui. Considere sempre anos completos, e que um ano possui 365 dias. Ex: uma pessoa com 19 anos possui 6935 dias de vida; veja um exemplo de saída: MARIA, VOCÊ JÁ VIVEU 6935 DIAS


```python
nome = input("Digite o seu nome:")
idade = int(input("Digite a sua idade:"))

qtde_dias = idade * 365

print(f"{nome}, Você já viveu {qtde_dias} dias")

dias_luta = qtde_dias * 0.973
dias_gloria = qtde_dias - dias_luta

print(f"Sendo {dias_luta} dias de luta e {dias_gloria} dias de gloria.")
```

5. Uma fábrica de camisetas produz os tamanhos pequeno, médio e grande, cada uma sendo vendida respectivamente por 10, 12 e 15 reais. Construa um algoritmo em que o usuário forneça a
quantidade de camisetas pequenas, médias e grandes referentes a uma venda, e a máquina informe quanto será o valor arrecadado.

```python
valor_cam_p = 10
valor_cam_m = 12
valor_cam_g = 15

qtde_cam_p, qtde_cam_m, qtde_cam_g = map(int, input("Digite 3 valores (ex: 5 10 2): ").split())

valor_total = (valor_cam_p * qtde_cam_p) + (valor_cam_m * qtde_cam_m) + (valor_cam_g * qtde_cam_g)

print(f"O valor total ficou em R${valor_total:.2f}")
```
